## 2. Carichiamo il dataset

Il file CSV contiene i dati delle 48 squadre qualificate al Mondiale 2026. E' lo stesso dataset che avete gia' usato su Google Sheets!

**Su Google Colab**: prima di eseguire la cella qui sotto, dovete caricare il file. Decommentate le due righe (togliete il `#`) e vi apparira' un bottone per scegliere il file dal vostro computer.

```python
# Decommentate queste 2 righe su Colab per caricare il file
# from google.colab import files
# uploaded = files.upload()

df = pd.read_csv('worldcup2026_teams.csv')
df.head(10)
```

`df.head(10)` ci mostra le prime 10 righe. Ogni riga e' una squadra, ogni colonna e' un'informazione (ranking, gol, punti, ecc.).

Quante squadre e quante colonne abbiamo in totale?

```python
# .shape ci dice (righe, colonne)
print(f'Il dataset ha {df.shape[0]} squadre e {df.shape[1]} colonne')
```

Perfetto. Vediamo anche l'elenco di tutte le colonne disponibili:

```python
list(df.columns)
```

---

## 3. Le prime statistiche

Ricordate media e mediana da Google Sheets? Qui facciamo la stessa cosa, ma con una riga di codice.

Partiamo dal **ranking FIFA**: un numero basso significa squadra forte (l'Argentina e' 1, il Brasile e' 6).

```python
print('Ranking FIFA medio:', df['fifa_ranking'].mean().round(1))
print('Ranking FIFA mediano:', df['fifa_ranking'].median())
```

La media e la mediana sono diverse? Se si', significa che ci sono valori estremi che "tirano" la media. Vi ricorda qualcosa? (L'esempio degli stipendi con il CEO del Cap 03!)

Ora guardiamo **gol e punti**:

```python
print('Gol segnati in media:', df['goals_for'].mean().round(1))
print('Gol subiti in media:', df['goals_against'].mean().round(1))
print('Punti medi:', df['points'].mean().round(1))
```

### Medie per confederazione

Domanda interessante: **quale confederazione sta andando meglio?** UEFA (Europa), CONMEBOL (Sud America), o qualcun altro?

Con `groupby` raggruppiamo le squadre per confederazione e calcoliamo la media dei punti:

```python
df.groupby('confederation')['points'].mean().sort_values(ascending=False).round(1)
```

Sorpresi? I numeri dicono sempre la verita', ma bisogna saperli leggere. Ricordatevi che alcune confederazioni hanno piu' squadre di altre, quindi la media potrebbe essere influenzata.

## 4. I grafici

I numeri raccontano una storia. I grafici la rendono **visibile**. (Come dicevamo nel Cap 04: visualizzare e' comunicare!)

### Grafico a barre: punti medi per confederazione

Prendiamo i numeri di prima e trasformiamoli in un grafico:

```python
media_conf = df.groupby('confederation')['points'].mean().sort_values(ascending=False)

media_conf.plot(kind='bar', color='#3C80C2', figsize=(8, 4))
plt.ylabel('Punti medi')
plt.title('Punti medi per confederazione')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Grafico a torta: squadre per continente

Quante squadre vengono dall'Europa? Dall'Africa? Dall'Asia? Un grafico a torta e' perfetto per mostrare le **proporzioni**:

```python
df['continent'].value_counts().plot(kind='pie', autopct='%1.0f%%', figsize=(6, 6))
plt.ylabel('')
plt.title('Squadre per continente')
plt.show()
```

### Grafico a dispersione: ranking FIFA vs punti

Questa e' la domanda da un milione di dollari: **le squadre con ranking migliore fanno davvero piu' punti?**

Usiamo uno scatter plot. Ogni pallino e' una squadra. Se i pallini seguono una tendenza (dal basso a sinistra verso l'alto a destra), allora si': ranking e punti sono collegati.

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['fifa_ranking'], df['points'], s=60, alpha=0.7, edgecolors='black', linewidth=0.5)
plt.xlabel('Ranking FIFA (piu basso = piu forte)')
plt.ylabel('Punti')
plt.title('Ranking FIFA vs Punti nel girone')
plt.gca().invert_xaxis()
plt.tight_layout()
plt.show()
```

Vedete una tendenza chiara o e' tutto sparso? Attenzione: anche se vedete una tendenza, ricordate la lezione sulla **correlazione** — non significa automaticamente che una cosa causa l'altra!

### Top 10 squadre per punti

Per finire, vediamo chi sta in cima alla classifica:

```python
top10 = df.nlargest(10, 'points')[['team', 'points', 'fifa_ranking', 'goals_for']]
top10.plot(kind='barh', x='team', y='points', color='#E7AF2A', figsize=(8, 4), legend=False)
plt.xlabel('Punti')
plt.title('Top 10 squadre per punti')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
